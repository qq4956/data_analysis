```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv('/Users/xiaruihan/Desktop/123.csv')
```

```python
df.head(5)
```

```python
df_clean = df[['saleprice','item_name','foreign_category_lv1_name','foreign_category_lv2_name','foreign_category_lv3_name']]
```

```python
df_clean.head()
```

```python
df_clean.info()
```

```python
df_item_price = df_clean.groupby('item_name').min()
```

```python
df_process = df_item_price.sort_values(by='saleprice',ascending = False)
df_out = df_process[:100]
```

```python
Y = df_out.saleprice.values
X = np.array(df_out.index)
Z = np.arange(0,len(Y)) 
    
```

```python
plt.figure(figsize = (50,10))
plt.bar(Z,Y)
```

```python
plt.xticks(Z,X)
```

```python
plt.xticks(rotation=90)
```

```python
ax = plt.gca()
from matplotlib.font_manager import FontProperties
font_zh = FontProperties(fname = '/Library/Fonts/Songti.ttc')
for label in ax.get_xticklabels():
    label.set_fontproperties(font_zh)
```
