#### 1.1 检查是否有重复值,并清除重复值

```python
len(df.positionId.unique())
```

```python
df_duplicates = df.drop_duplicates(subset = 'positionId')
```

#### 1.2 清洗薪资字段

```python
def cut_word(word,method = 'top'):
    position = word.find('-')
    length = len(word)
    if position != -1:
        bottomSalary = word[:position-1]
        topSalary = word[position+1:length-1]
    else:
        position = word.upper().find('K')
        bottomSalary = word[:position]
        topSalary =word[:position]
    if method == 'bottom':
        return bottomSalary
    else:
        return topSalary
```

```python
df_duplicates['topSalary'] = df_duplicates.salary.apply(cut_word)
df_duplicates['bottomSalary'] = df_duplicates.salary.apply(cut_word,method='bottom')

df_duplicates
```

```python
#转换数据格式
```

```python
df_duplicates.topSalary = df_duplicates.topSalary.astype('int')
df_duplicates.bottomSalary = df_duplicates.bottomSalary.astype('int')
```

```python
## 得到平均薪资字段
```

```python
df_duplicates['avgSalary'] = df_duplicates.apply(lambda x :(x.topSalary + x.bottomSalary)/2,axis = 1)
```

### 2.提取已经清洗干净的数据

```python
df_clean = df_duplicates[['city','companySize','education','companyShortName','industryField','positionName','positionLables','workYear','avgSalary']]
```

### 3.5 分析北京上海分组的薪资水平

```python
bins = [0,3,5,10,13,15,20,30,100]
level = ['0-3','3-5','5-10','10-13','13-15','15-20','20-30','30+']
df_clean['level'] = pd.cut(df_clean['avgSalary'],bins = bins,labels = level)
```

```python
df_level = df_clean.groupby(['city','level']).avgSalary.count().unstack()
df_level_prop = df_level.apply(lambda x:x/x.sum(),axis = 1)
ax = df_level_prop.plot.bar(stacked = True ,figsize=(20,10))
for label in ax.get_xticklabels():
    label.set_fontproperties(font_zh)
ax.legend(prop = font_zh)
```
