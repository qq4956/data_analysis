```python
import pandas as pd
import numpy as np
```

```python
goodsflow = pd.read_csv('/Users/xiaruihan/Desktop/天天cc品探索/天天12月数据.csv')
inventory = pd.read_csv('/Users/xiaruihan/Desktop/天天cc品探索/tt_tmp_inventory_1.2.csv',encoding = 'gb2312')
goods = pd.read_csv('/Users/xiaruihan/Desktop/天天cc品探索/chain_goods.csv')
store = pd.read_csv('/Users/xiaruihan/Desktop/天天cc品探索/foreign_store_id.csv')
```

```python
%%timeit
store_list = np.array(store).tolist()
for x in store_list:
    goodsflow_x = goodsflow[goodsflow.foreign_store_id == x ]
    goodsflow_x = goodsflow_x.foreign_item_id.unique()
    goodsflow_x = pd.DataFrame(goodsflow_x)
    goodsflow_x.columns = ['foreign_item_id']
    inventory_x = inventory[['foreign_item_id','foreign_store_id','quantity']]
    inventory_x = inventory_x[inventory_x.foreign_store_id == x]
    inventory_index = inventory_x.groupby('foreign_item_id').sum()
    inventory_index = inventory_index[inventory_index.quantity != 0]
    inventory_index = np.array(inventory_index.index)
    inventory_y = pd.DataFrame(inventory_index)
    inventory_y.columns = ['foreign_item_id']
    res = pd.merge(goodsflow_x, inventory_y, on=['foreign_item_id', 'foreign_item_id'])
    goodsflow_num = len(goodsflow_x)
    inventory_num = len(inventory_y)
    res_num = len(res)
    diff = inventory_num - res_num
    print('店铺编号:',x,'流水商品数:',goodsflow_num,'库存数:',inventory_num,'有销售有库存数:',res_num,'无销售有库存数:',diff)
```
